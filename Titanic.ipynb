# Load and Verify the Data

```python
import pandas as pd
titanic = pd.read_csv('Titanic.csv')
```

```python
titanic.head()
```

# Understand the Features

###### PassengerId:  A unique index for each passenger.
###### Survived: Shows if the passenger survived or not (1 = survived, 0 = did not survive).
###### Pclass: Passenger Class (1 = First class ticket, 2 = Second class ticket, 3 = Third class ticket).
###### Name: The passenger’s name, which also contains the title (Mr., Mrs., Doctor, et cetera)
###### Sex: The passenger’s gender (male or female).
###### Age: The passenger’s age (or if the age was not recorded, NaN).
###### SibSp: The number of siblings and spouses traveling with that passenger.
###### Parch: The number of parents and children traveling with that passenger.
###### Ticket: The passenger’s ticket number.
###### Fare: The price of the passenger's ticket.
###### Cabin: The passenger’s cabin number (or if the cabin number was not recorded, Nan).
###### Embarked: The port where the passenger boarded (C = Cherbourg, Q = Queenstown, S = Southampton)

# Impute Missing Data

###### Fill in the missing Age and Fare values (both float64 data types) by imputing the median (which is more robust to outliers).

```python
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())
titanic['Fare'] = titanic['Fare'].fillna(titanic['Fare'].median())
```

# Drop Unnecessary Columns

```python
titanic = titanic.drop(titanic[['PassengerId', 'Name', 'Ticket', 'Fare', 'Cabin']], axis = 1)
```

# Drop Unnecessary Rows

```python
titanic = titanic.drop(titanic[titanic['Embarked'].isna()].index)
```

###### Check again for missing data

```python
titanic.isnull().sum()
```

```python
titanic.info()
```

# Mapping

###### There are still two columns (Sex and Embarked) that contain non-numeric values. Those non-numeric values will have to be mapped to different numbers.  First, obtain a list of unique values in each column.

```python
titanic['Sex'].unique()
```

```python
titanic['Embarked'].unique()
```

```python
titanic['Sex'] = titanic['Sex'].map({'male':0, 'female':1})
titanic['Embarked'] = titanic['Embarked'].map({'C':0, 'Q':1, 'S':2})
```

```python
titanic.head()
```

# Separate the Features from the Targets

```python
features = titanic.drop(['Survived'], axis = 1)
targets = titanic['Survived']
```

# Split into Train and Test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, targets, test_size = 0.34, random_state = 42)
```

# Create a Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

# Train the Model

```python
model.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test)
```

# Evaluate the Model Performance

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, prediction))
```
